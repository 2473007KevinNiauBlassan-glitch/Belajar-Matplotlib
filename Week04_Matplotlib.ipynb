```python
#Bar Plot
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
     
```

```python

#Simple Bar plot
data=[25,45,55,125,225]
kategori = ['A','B','C','D','E',]
```

```python

plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.tittle('Contoh Bar Plot')

plt.show()
```

```python
#Pengaturan Grid dan Color
data=[25,45,55,125,225]
kategori = ['A','B','C','D','E']
     
```

```python
plt.bar(kategori,
        data,
        color='red',
        alpha=0.25)

plt.grid(linestyle='--',
         linewidth=2,
         axis='y', #pilihan:x,y,both
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('JUmlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Horisontal Bar Plot
data=[25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.barh(kategori, data)
plt.grid(linestyle='--',
         linewidth=1,
         axis='x', 
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('JUmlah')
plt.title('Contoh Horisontal Bar Plot')
plt.show
```

```python
plt.barh(kategori, data)
plt.barch(kategori[::-1], data[::-1])
plt.grid(linestyle='--',
         linewidth=1,
         axis='x', 
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('JUmlah')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
     
```

```python

#Stacked Bar Plot
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori = ['A','B','C','D','E']
     
```

```python
plt.bar(kategori, data1, label='Data 1')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='x', 
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('JUmlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python

plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data1, label='Data 1')

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='x', 
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('JUmlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
#Simple Pie Plot
kategori = ['A','B','C','D','E']
data=[25,45,55,125,225]
```

```python
plt.pie(data, 
        labels=kategori,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python

#Pengaturan Legend dan Color
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
     
```

```python

plt.pie(data, 
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
#Exploded Pie Plot
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
     
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Exploded Pie Plot')

plt.show()
     
```
